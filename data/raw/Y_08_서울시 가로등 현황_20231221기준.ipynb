```python
import pandas as pd
from geopy.geocoders import Nominatim
import time
from tqdm import tqdm 
import os # (파일 경로 관리를 위해 os 임포트)

tqdm.pandas()

# --- 1. CSV 파일 로드 ---
base_path = r'C:\Users\user\Desktop\경진대회'
file_name = 'Y_08_서울시 가로등 현황_20231221기준.csv'
file_path = os.path.join(base_path, file_name)

try:
    df = pd.read_csv(file_path, encoding='euc-kr') 
except FileNotFoundError:
    print(f"'{file_path}' 파일을 찾을 수 없습니다.")
    exit() 
except Exception as e:
    print(f"CSV 파일 읽기 오류: {e}")
    exit() 

print(f"1. CSV 파일 로드 완료. (총 {len(df)}건)")

# --- (수정) 전체 데이터로 실행 ---
print("2. ⭐️ 전체 데이터로 실행을 준비합니다. ⭐️")
# df_sample = df.head(10).copy() # (테스트용)
 
# ⭐️⭐️⭐️ 전체 데이터로 실행하도록 수정 ⭐️⭐️⭐️
df_full = df.copy() 

print(f"(총 {len(df_full)}건의 데이터로 역지오코딩을 시작합니다.)")
print(f"⚠️ 경고: 예상 소요 시간 약 {len(df_full) * 1.1 / 3600 :.2f} 시간")
print("⚠️ 절대로 중간에 멈추거나 프로그램을 끄지 마세요!")

# --- 2. 지오코더 설정 ---
geolocator = Nominatim(user_agent="my-reverse-geocoder-app-v3")

# --- 3. (최종) '구' 이름을 가져오는 함수 ---
def get_gu_from_coords(row):
    try:
        coords = (row['위도'], row['경도'])
        time.sleep(1.1) # (필수) 1.1초 대기
        
        location = geolocator.reverse(coords, language='ko', timeout=10)
        
        if location and location.raw.get('address'):
            address = location.raw['address']
            
            # 1순위: 'borough' 필드 확인 (e.g., '송파구')
            gu = address.get('borough')
            if gu and '구' in gu:
                return gu

            # 2순위: 'city_district' 필드 확인
            gu = address.get('city_district')
            if gu and '구' in gu:
                return gu
            
            # 3순위: 'county' 필드 확인
            gu = address.get('county')
            if gu and '구' in gu:
                return gu
            
            return '기타 (구 정보 없음)'
        else:
            return '주소 없음'
    except Exception as e:
        # 오류가 나도 멈추지 않고 '오류'라고 기록
        return f'오류' 

# --- 4. 함수 적용 (전체 데이터) ---
print(f"3. 역지오코딩을 시작합니다...")
df_full['지역구'] = df_full.progress_apply(get_gu_from_coords, axis=1)

# --- 5. 결과 확인 및 집계 ---
print("\n--- 역지오코딩 완료 ---")
# (결과가 너무 많으므로 상위 5개만 샘플로 출력)
print(df_full[['관리번호', '위도', '경도', '지역구']].head(5))

print("\n--- 지역구별 가로등 개수 집계 (전체) ---")
gu_counts = df_full['지역구'].value_counts()
print(gu_counts)

# --- 6. (⭐️ 신규 ⭐️) CSV 파일로 저장 ---
output_filename = '서울시_가로등_현황_지역구_포함.csv'
output_path = os.path.join(base_path, output_filename)

try:
    # 'euc-kr' 인코딩으로 저장해야 Excel에서 한글이 깨지지 않습니다.
    # index=False로 설정하여 불필요한 인덱스 번호를 저장하지 않습니다.
    df_full.to_csv(output_path, encoding='euc-kr', index=False)
    
    print("\n--- 7. 파일 저장 완료! ---")
    print(f"축하합니다! 모든 작업이 완료되었습니다.")
    print(f"경로: {output_path}")

except Exception as e:
    print(f"\n--- 7. 파일 저장 중 오류 발생 ---")
    print(f"오류: {e}")
```
