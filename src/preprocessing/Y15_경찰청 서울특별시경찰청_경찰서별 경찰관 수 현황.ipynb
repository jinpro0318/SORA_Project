```python
import pandas as pd
file_name = "Y15_경찰청 서울특별시경찰청_경찰서별 경찰관 수 현황_20241231.csv"
df_original = pd.read_csv(f"./{file_name}",encoding="euc-kr")
df = df_original.copy()
df.head()
```

```python
df.columns = [col.replace('년', '') for col in df.columns]
df.head()
```

```python
df = df[['경찰서'] + [str(y) for y in range(2020, 2025)]]
df = df.sort_values(by='2024', ascending=False).reset_index(drop=True)
df
```

```python
# 컬럼명 변경
df.rename(columns={'경찰서': '지역구'}, inplace=True)

# 지역구 이름에 '구' 붙이기
df['지역구'] = df['지역구'].astype(str) + '구'

# 지역구 기준 오름차순 정렬
df = df.sort_values(by='지역구').reset_index(drop=True)

# 결과 확인
print(df)
```

```python
# csv파일로 저장
file_name = 'Y15_전처리_경찰관_인원수_2020_2024.csv'

# DataFrame을 CSV 파일로 저장합니다.
df.to_csv(file_name, encoding='utf-8', index=False)

print(f"DataFrame이 '{file_name}' 이름으로 저장되었습니다.")
```
