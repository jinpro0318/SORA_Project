```python
import pandas as pd

file_name = "X13_독거노인현황2.csv"

df_original = pd.read_csv(
    f"./{file_name}",
    encoding="utf-8",
    engine="python",       # C 엔진 대신 Python 엔진 사용
    on_bad_lines="skip",   # 깨진 줄 무시
)

df = df_original.copy()
df.head(15)
```

```python
df.columns.tolist()
```

```python
df = df.drop(index=[0, 1, 2]).reset_index(drop=True)
df
```

```python
# 2. 컬럼명을 다시 한번 정리합니다.
df.columns = df.columns.str.strip()

df = df.drop(df.columns[0], axis=1)

df.rename(columns={'동별(2)': '지역구', '동별(3)': '동별'}, inplace=True)
print(len(df))
df.head(15)
```

```python
subtotal_df = df[df['동별'] == '소계'].copy()
# 1. 첫 번째 행 (인덱스 0) 삭제
subtotal_df.drop(index=0, inplace=True)
# 2. '동별' 컬럼 삭제
subtotal_df.drop(columns='동별', inplace=True)
# 3. '지역구' 컬럼을 기준으로 오름차순 정렬
subtotal_df.sort_values(by='지역구', ascending=True, inplace=True)
subtotal_df.reset_index(drop=True, inplace=True)
subtotal_df
```
