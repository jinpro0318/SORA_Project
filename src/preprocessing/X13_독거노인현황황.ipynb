```python
import pandas as pd

file_name = "X13_독거노인현황2.csv"

df_original = pd.read_csv(
    f"./{file_name}",
    encoding="utf-8-sig",
    engine="python",       # ✅ C 엔진 대신 Python 엔진 사용
    on_bad_lines="skip",   # 깨진 줄 무시
)

df = df_original.copy()
df.head(15)
```

```python
df = df.drop(index=[0, 1, 2]).reset_index(drop=True)
df
```

```python
# 2. 컬럼명을 다시 한번 정리합니다.
df.columns = df.columns.str.strip()

# 3. '동별(1)' 컬럼을 삭제합니다.
#    '동별(1)'이 첫 번째 컬럼(인덱스 0)일 가능성이 높으므로,
#    이름 대신 인덱스를 사용하여 가장 안전하게 삭제합니다.
#    *만약 출력된 컬럼명에 '동별(1)'과 비슷한 이름이 있다면, 그 이름을 사용해야 합니다.*
#    **가장 확실한 방법: 첫 번째 컬럼을 삭제**
df = df.drop(df.columns[0], axis=1)

# 4. 남은 컬럼명 수정 ('동별(2)'와 '동별(3)'도 이름에 띄어쓰기 등이 남아있을 수 있으므로 strip() 후 수정)
df.rename(columns={'동별(2)': '지역구', '동별(3)': '동별'}, inplace=True)
df.head(50)
```
