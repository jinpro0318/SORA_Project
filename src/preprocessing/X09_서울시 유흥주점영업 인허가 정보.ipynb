```python
df.columns = df.columns.str.strip()

# ⚠️ '개방자치단체코드' 컬럼이 문자열 형태인지 확인
# 만약 숫자로 되어 있다면 문자열로 변환해야 매핑이 정확합니다.
df['개방자치단체코드'] = df['개방자치단체코드'].astype(str)

# 서울시 25개 자치구의 표준 매핑 딕셔너리
seoul_code_to_gu = {
    '3000000': '종로구', '3010000': '중구', '3020000': '용산구', '3030000': '성동구',
    '3040000': '광진구', '3050000': '동대문구', '3060000': '중랑구', '3070000': '성북구',
    '3080000': '강북구', '3090000': '도봉구', '3100000': '노원구', '3110000': '은평구',
    '3120000': '서대문구', '3130000': '마포구', '3140000': '양천구', '3150000': '강서구',
    '3160000': '구로구', '3170000': '금천구', '3180000': '영등포구', '3190000': '동작구',
    '3200000': '관악구', '3210000': '서초구', '3220000': '강남구', '3230000': '송파구',
    '3240000': '강동구'
}
# '개방자치단체코드' 컬럼에 딕셔너리를 직접 매핑하여 '지역구명' 컬럼을 생성합니다.

df['지역구명'] = df['개방자치단체코드'].map(seoul_code_to_gu)
df['지역구명']
```

```python
df['상세영업상태명'] 
df = df[df['상세영업상태명'].str.strip() == '영업']
df
```

```python
len(df['상세영업상태명'])
```

```python

# 1. '인허가일자'를 날짜/시간(datetime) 형식으로 변환합니다.
#    errors='coerce': 변환 불가능한 값은 NaT(Not a Time)로 처리합니다.
df.columns = df.columns.str.strip()
df['인허가일자'] = pd.to_datetime(df['인허가일자'], errors='coerce')

# 2. .dt.year 속성을 사용하여 년도 4자리만 추출합니다.
df['인허가년도'] = df['인허가일자'].dt.year

df['인허가년도']
```

```python
df['지번주소'] = df['지번주소'].astype(str)
df['동별'] = df['지번주소'].str.split(' ').str.get(2)
df['동별']

#df['동별'].isnull().sum()
```

```python
df = df[['인허가년도', '지역구명', '동별']]
df = df.sort_values(by='인허가년도', ascending=True).reset_index(drop=True)
df
```

```python
#df.isnull().sum()

df_missing_dong = df[df['동별'].isnull()]
df = df.reset_index(drop=True)
len(df)
```

```python
# '지역구명'별 유흥주점 수
df_gu_count = df.groupby('지역구명').size().reset_index(name='지역구별 주점 수')

# 유흥주점 수 기준으로 내림차순 정렬하여 순위를 확인합니다.
df_gu_count = df_gu_count.sort_values(by='지역구별 주점 수', ascending=False).reset_index(drop=True)
print(df_gu_count)

total_gu_sum = df_gu_count['지역구별 주점 수'].sum()
total_gu_sum

```

```python
# '지역구명'별 유흥주점 수
df_gu_count = df.groupby('동별').size().reset_index(name='동별 주점 수')

# 유흥주점 수 기준으로 내림차순 정렬하여 순위를 확인
df_gu_count = df_gu_count.sort_values(by='동별 주점 수', ascending=False).reset_index(drop=True)
print(df_gu_count)

total_gu_sum = df_gu_count['동별 주점 수'].sum()
total_gu_sum
```

```python
# 1. 동별 결측치 제거 (이전 단계에서 최종적으로 결정된 처리)
df = df.dropna(subset=['동별']).copy()
df = df.reset_index(drop=True)

# 2. 지역구별 유흥주점 총합 계산
# '지역구명'을 기준으로 그룹화하여 전체 개수를 셉니다.
df['지역구별 주점 수'] = df.groupby('지역구명')['동별'].transform('size')

# 3. 지역구 내 동별 유흥주점 합계 계산
# '지역구명'과 '동별'을 함께 기준으로 그룹화하여 개수를 셉니다.
df['동별 주점 수'] = df.groupby(['지역구명', '동별'])['동별'].transform('size')


# 4. 최종 결과 테이블 생성
# 필요한 컬럼만 남기고 중복 제거 (각 지역구/동 조합당 하나의 행만 남김)
df_result = df[['지역구명', '동별', '지역구별 주점 수', '동별 주점 수']].drop_duplicates().reset_index(drop=True)

# 5. 보기 좋게 정렬
# 지역구 총합 내림차순 -> 지역구명 오름차순 -> 동별 주점 수 내림차순
df_result = df_result.sort_values(
    by=['지역구별 주점 수', '지역구명', '동별 주점 수'],
    ascending=[False, True, False]
).reset_index(drop=True)


df_result
```

```python
# csv 파일로 데이터 저장
file_name = "X09_전처리_서울시 유흥주점.csv"

df_result.to_csv(
    file_name,
    index=False,
    encoding='cp949' 
)

print(f"✔️ 데이터가 '{file_name}'으로 성공적으로 저장되었습니다.")
```
